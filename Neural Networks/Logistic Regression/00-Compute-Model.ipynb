```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([10, 2, 18, 24, 11, 32, 54, 35, 67, 4, 16, 17])
y = np.array([0, 0, 1, 1, 0, 1, 1, 1, 1, 0, 0, 0])

len(X), len(y)
```

```python
# plot data


plt.figure(figsize = (15, 5))
plt.scatter(X, y, marker = 'x')
```

```python
def compute_linear(X, w, b):

    return w * X + b    
```

```python
def compute_model(X, w, b):

    z = compute_linear(X, w, b)

    exp = np.exp(-z)

    g_z = 1 / ( 1 + exp)

    return g_z
```

```python
y_pred = compute_model(X, -10, -30);

y_pred
```
