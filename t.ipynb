```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data set

x = np.array([0, 5, 10, 15, 20])
y = np.array([5, 10, 15, 20, 25])
y_pred = np.array([10, 15, 20, 25, 30])
```

```python
m = 5

for i in range(m):

    y_pred -= 1
    plt.scatter(x, y, marker = 'x', c = 'r', label = 'Original')
    pred_line = plt.plot(x, y_pred, c = 'b', label = 'Predictions')

    p = pred_line.pop(0)
    p.remove()
```
