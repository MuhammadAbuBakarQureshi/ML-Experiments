```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create Data set

X = np.array([10, 2, 19, 24, 11, 32, 54, 35, 67, 4, 16, 17])
y = np.array([0, 0, 1, 1, 0, 1, 1, 1, 1, 0, 0, 0])

len(X), len(y)
```

```python
# plot data set

plt.figure(figsize = (15, 5))
plt.scatter(X, y, marker = 'x', c = 'red')
```

```python
def compute_z(X, w, b):

    return w * X + b
```

```python
def compute_model(X, w, b):

    z = compute_z(X, w, b)

    exp = np.exp(-z)

    g_z = 1 / ( 1 + exp)

    return g_z
```

```python
def compute_cost(X, y, w, b):

    f_wb = compute_model(X, w, b)

    cost = (-y * ( np.log(f_wb))) - ( (1 - y) * np.log(1 - f_wb) )

    cost = np.sum(cost)/ len(X)

    return cost
```

```python
y_pred = compute_model(X, 5, -90)
```

```python
# before threshold

y_pred
```

```python
threshold = 0.5

for i in range(len(y_pred)):

    if (y_pred[i] > threshold):
        
        y_pred[i] = 1

    else:

        y_pred[i] = 0
```

```python
# after threshold

y_pred
```

```python
# plot the predictions

plt.figure(figsize = (15, 5))
plt.scatter(X, y, marker = 'o', c = 'black')
plt.scatter(X, y_pred, marker = 'x', c = 'red')
plt.legend()
```
